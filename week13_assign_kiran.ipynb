Problem Statement1:
You are tasked with building a database for an e-commerce system. Create a database named ecommerce.db and add a table named Customers to store details about customers. The table should include:

* customer_id (INTEGER): A unique ID for each customer.
* name (TEXT): The full name of the customer.
* email (TEXT): The email address of the customer.
* created_at (DATE): The date when the customer signed up.

Expected Output:Database created, and Customers table created successfully!


```python
import sqlite3
connection=sqlite3.connect("ecommerce.db")
cursor = connection.cursor()
customers_table_query = """
CREATE TABLE Customers (
    customer_id INTEGER PRIMARY KEY,
    name TEXT,
    email TEXT,
    created_at DATE
);
"""
cursor.execute(customers_table_query)
connection.commit()
print("Database created, and Customers table created successfully!")
```

Problem Statement2:
Extend the ecommerce.db database by adding another table named Orders. This table will store details about customer orders, including:

* order_id (INTEGER): A unique ID for each order.
* customer_id (INTEGER): The ID of the customer who placed the order (linked to the Customers table).
* order_date (DATE): The date when the order was placed.
* amount (INTEGER): The total amount of the order.

Expected Output:Orders table created successfully!

```python
orders_table_query = """
CREATE TABLE Orders (
    order_id INTEGER PRIMARY KEY,
    customer_id INTEGER,
    order_date DATE,
    amount INTEGER,
    FOREIGN KEY (customer_id) REFERENCES Customers(customer_id)
);
"""
cursor.execute(orders_table_query)
connection.commit()
print("Orders table created successfully!")
```
