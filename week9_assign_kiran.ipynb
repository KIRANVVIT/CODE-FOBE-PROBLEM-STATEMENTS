1.Problem Statement: Create two Series, representing the scores of two exams, and add them together.

2.Problem Statement:
Create a Pandas Series containing the daily temperatures for five days. Filter and display only the temperatures greater than 30°C.

temperatures: [28, 35, 31, 29, 25]

Index: ["Day 1", "Day 2", "Day 3", "Day 4", "Day 5"]

Output:

Temperatures Greater than 30°C:

Day 2    35

Day 3    31

dtype: int64

```python
#creating pandas series daily temp 5 days, filter temp > 30 degree celcius
import pandas as pd
temperatures = pd.Series([28, 35, 31, 29, 25], index=["Day 1", "Day 2", "Day 3", "Day 4", "Day 5"])
temp_greater_30 = temperatures[temperatures > 30]
print("Temperatures Greater than 30°C:\n",temp_greater_30)
```

3.Problem Statement:
Create a Pandas Series containing five values, including two missing values represented as NaN. Replace all missing values with 0.

data: [10, NaN, 20, NaN, 30]

Output:

Series with Missing Values Replaced:

0    10.0

1     0.0

2    20.0

3     0.0

4    30.0

dtype: float64

```python
#Series containing 5 values, 2 missing values as NaN. Replace missing values with 0.data: [10, NaN, 20, NaN, 30]
import pandas as pd
import numpy as np
data = pd.Series([10, np.nan, 20, np.nan, 30])
data[pd.isnull(data)]=0
print("Series with Missing Values Replaced:\n",data)
```
