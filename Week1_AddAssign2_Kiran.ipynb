Q2.Convert total minutes to hours and minutes

Take a total number of minutes from the user and display it as hours and remaining minutes.




Expected Output

Q3. Simple salary calculation

Ask for employee name, hourly wage, and hours worked. Calculate weekly pay and print nicely.


Expected Output

Employee name: alisha

Hourly wage: 200

Hours worked: 10

alisha's weekly pay is ₹2000.0


```python
#enter your name
name=input("Employee name: ")
#enter hourly wage hrwg
hrwg=int(input("Hourly wage: "))
#enter hours worked hrsworked
hrsworked=int(input("Hours worked: "))
#calculating weekly pay
weeklypay=hrwg*hrsworked
print(f"{name}'s weekly pay is {float(weeklypay)}")

```

Q6. Simple discount calculator

Ask for original price and discount percentage. Display final price rounded to 2 decimal places.

Expected Output

Enter the original price: 400

Enter discount percentage: 10

Final price after 10.0% discount = ₹360.00

```python
#enter original price op
op=int(input("Enter the original price: "))
#enter discound percentage dp
dp=int(input("Enter discount percentage: "))
#calculating final price fp
fp=op-(op*dp/100)
print(f"Final price after {float(dp)}% discound = {float(fp)}")
```
