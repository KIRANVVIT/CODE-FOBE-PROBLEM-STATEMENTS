Problem 2: Box Plot – Customized

Using the same data as Problem 1, customize the box plot with colors, labels, and gridlines.

```python
import matplotlib.pyplot as plt
test_scores = [45, 55, 60, 62, 65, 70, 72, 75, 80, 85, 90, 92, 95, 100, 60, 70, 75]

# Customizing the box plot
plt.boxplot(test_scores, patch_artist=True,
            boxprops=dict(facecolor="red", color="yellow"),
            whiskerprops=dict(color="orange", linestyle="--"),
            capprops=dict(color="darkgreen"),
            medianprops=dict(color="yellow")
            )
plt.title("Test Score Distribution", fontsize=16, color='aqua',fontweight='bold')
plt.ylabel("Scores", fontsize=12, color='magenta')
plt.grid(axis="y", linestyle="--", alpha=0.7, color='black')
plt.show()
```
