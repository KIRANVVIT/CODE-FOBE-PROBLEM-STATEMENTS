✅ Problem Statement6: Create an POST API to Add a New Book to the Database

````python
from flask import Flask, jsonify, request
import sqlite3

app = Flask(__name__)

# Home Route
@app.route('/')
def home():
    return jsonify({
        "message": "Welcome to the Online Bookstore API. Use '/books' to fetch all books, '/books/<id>' to fetch a book, or POST to '/books' to add a new book."
    })

# GET API - Fetch All Books
@app.route('/books', methods=['GET'])
def get_books():
    conn = sqlite3.connect('books.db')
    cursor = conn.cursor()

    cursor.execute("SELECT * FROM books")
    books = cursor.fetchall()

    conn.close()

    books_list = []
    for book in books:
        books_list.append({
            "id": book[0],
            "title": book[1],
            "author": book[2],
            "price": book[3],
            "stock_quantity": book[4]
        })

    return jsonify(books_list)

# GET API - Fetch Book by ID
@app.route('/books/<int:book_id>', methods=['GET'])
def get_book(book_id):
    conn = sqlite3.connect('books.db')
    cursor = conn.cursor()

    cursor.execute("SELECT * FROM books WHERE id=?", (book_id,))
    book = cursor.fetchone()

    conn.close()

    if book:
        return jsonify({
            "id": book[0],
            "title": book[1],
            "author": book[2],
            "price": book[3],
            "stock_quantity": book[4]
        })
    else:
        return jsonify({"error": "Book not found"}), 404

# POST API - Add New Book
@app.route('/books', methods=['POST'])
def add_book():
    data = request.get_json()

    title = data.get('title')
    author = data.get('author')
    price = data.get('price')
    stock_quantity = data.get('stock_quantity')

    # Validate Input Data
    if not all([title, author, price, stock_quantity]):
        return jsonify({
            "error": "Please provide title, author, price, and stock_quantity"
        }), 400

    conn = sqlite3.connect('books.db')
    cursor = conn.cursor()

    cursor.execute(
        "INSERT INTO books (title, author, price, stock_quantity) VALUES (?, ?, ?, ?)",
        (title, author, price, stock_quantity)
    )

    conn.commit()
    conn.close()

    return jsonify({
        "message": "Book added successfully!"
    }), 201

# Run the Flask App
if __name__ == '__main__':
    app.run(host='0.0.0.0', port = 5000,debug=True)
````

✅ Problem Statement7: Create an API to Fetch and Update a Book by ID

Develop an online bookstore API that allows users to retrieve and update book details using a unique ID.

* 1️⃣ GET /books/<id> → Fetch book details (title, author, price, stock_quantity).
* 2️⃣ PUT /books/<id> → Update book details by sending a JSON request.

✅ Step 1: Import Libraries
* Import Flask, sqlite3, and request for API handling.

✅ Step 2: Initialize Flask App
* Create app = Flask(__name__).

✅ Step 3: Define Home Route (/)

Display API instructions:

✅ Step 4: Define GET API (/books/<id>)
* Retrieve book details by ID and Return error message if the book is not found.

✅ Step 5: Define PUT API (/books/<id>)
* Accept JSON input and validate required fields.

````python
'''Create an API to Fetch and Update a Book by ID
Develop an online bookstore API that allows users to retrieve and update book details using a unique ID.
1️⃣ GET /books/ → Fetch book details (title, author, price, stock_quantity).
2️⃣ PUT /books/ → Update book details by sending a JSON request.
'''
from flask import Flask, jsonify, request
import sqlite3

# Initialize Flask App
app = Flask(__name__)

# Home Route
@app.route('/')
def home():
    return jsonify({
        "message": "Use '/books/<id>' to fetch a book and PUT to '/books/<id>' to update book details."
    })

# GET API - Fetch Book by ID
@app.route('/books/<int:book_id>', methods=['GET'])
def get_book(book_id):
    conn = sqlite3.connect('books.db')
    cursor = conn.cursor()

    cursor.execute("SELECT * FROM books WHERE id=?", (book_id,))
    book = cursor.fetchone()

    conn.close()

    if book:
        return jsonify({
            "id": book[0],
            "title": book[1],
            "author": book[2],
            "price": book[3],
            "stock_quantity": book[4]
        })
    else:
        return jsonify({
            "error": "Book not found"
        }), 404

# PUT API - Update Book by ID
@app.route('/books/<int:book_id>', methods=['PUT'])
def update_book(book_id):
    data = request.get_json()

    title = data.get('title')
    author = data.get('author')
    price = data.get('price')
    stock_quantity = data.get('stock_quantity')

    # Validate Input
    if not all([title, author, price, stock_quantity]):
        return jsonify({
            "error": "Please provide title, author, price, and stock_quantity"
        }), 400

    conn = sqlite3.connect('books.db')
    cursor = conn.cursor()

    # Check if book exists
    cursor.execute("SELECT * FROM books WHERE id=?", (book_id,))
    book = cursor.fetchone()

    if not book:
        conn.close()
        return jsonify({
            "error": "Book not found"
        }), 404

    # Update book
    cursor.execute("""
        UPDATE books
        SET title=?, author=?, price=?, stock_quantity=?
        WHERE id=?
    """, (title, author, price, stock_quantity, book_id))

    conn.commit()
    conn.close()

    return jsonify({
        "message": "Book updated successfully!"
    })

# Run Flask App
if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000, debug=True)
````

Develop an online bookstore API that allows users to retrieve all books and delete a specific book using its unique ID.

✅ Step 1: Import Necessary Libraries

* Retrieve all books and return them in JSON format.

✅ Step 6: Define Function to Check if a Book Exists
* Validate if a book exists in the database before fetching or deleting it.

✅ Step 7: Define DELETE API (/books/<id>)
* Check if the book exists before deleting it.
* If found, delete it and return a success message.
* If not found, return an error message (404 Book not found).

✅ Step 8: Run the Flask App
* Use 0.0.0.0:8080 for Replit compatibility.

**Expected Output**

````python
#Fetch All Books and Delete a Book by ID
from flask import Flask, jsonify
import sqlite3

# Initialize Flask App
app = Flask(__name__)

# Home Route
@app.route('/')
def home():
    return jsonify({
        "message": "Use '/books' to fetch all books or '/books/<id>' to delete a book."
    })

# Function to Fetch All Books
def fetch_all_books():
    conn = sqlite3.connect('books.db')
    cursor = conn.cursor()

    cursor.execute("SELECT * FROM books")
    books = cursor.fetchall()

    conn.close()
    return books

# Function to Check if a Book Exists
def book_exists(book_id):
    conn = sqlite3.connect('books.db')
    cursor = conn.cursor()

    cursor.execute("SELECT * FROM books WHERE id=?", (book_id,))
    book = cursor.fetchone()

    conn.close()
    return book

# GET API - Fetch All Books
@app.route('/books', methods=['GET'])
def get_books():
    books = fetch_all_books()

    books_list = []
    for book in books:
        books_list.append({
            "id": book[0],
            "title": book[1],
            "author": book[2],
            "price": book[3],
            "stock_quantity": book[4]
        })

    return jsonify(books_list)

# DELETE API - Delete Book by ID
@app.route('/books/<int:book_id>', methods=['DELETE'])
def delete_book(book_id):

    if not book_exists(book_id):
        return jsonify({
            "error": "Book not found"
        }), 404

    conn = sqlite3.connect('books.db')
    cursor = conn.cursor()

    cursor.execute("DELETE FROM books WHERE id=?", (book_id,))
    conn.commit()
    conn.close()

    return jsonify({
        "message": f"Book with ID {book_id} deleted successfully!"
    })

# Run Flask App
if __name__ == '__main__':
    app.run(host='0.0.0.0',port=5000,debug=True)
````

✅ Problem Statement9: Password Strength Checker API with Home Route

Users often create weak passwords, making their accounts vulnerable to hacking. This API helps users check the strength of their passwords and provides a rating (Weak, Medium, or Strong) based on certain rules.

🔹 API Functionalities:

1️⃣ Home Route (/) – Displays API instructions.

2️⃣ Password Strength Checker (/password-check/<password>) –

* Weak: If the password has less than 6 characters.
* Medium: If the password is 6-10 characters long and contains only letters and numbers.
* Strong: If the password is more than 10 characters and contains a mix of letters, numbers, or special characters.

🔹 Use Case:

Helps users create strong passwords for better security.
Provides a quick password strength assessment.

✅ Now, the API allows users to check their password security level! 🚀

**Expected Output**

image.png



image.png

````python
# Password Strength Checker API with Home Route
from flask import Flask, jsonify
import re

# Initialize Flask App
app = Flask(__name__)

# Home Route
@app.route('/')
def home():
    return jsonify({
        "message": "Welcome to the Password Strength Checker API. Use '/password-check/<password>' to check password strength."
    })

# Password Strength Checker API
@app.route('/password-check/<password>', methods=['GET'])
def check_password(password):

    if len(password) < 6:
        strength = "Weak"

    elif 6 <= len(password) <= 10 and password.isalnum():
        strength = "Medium"

    elif len(password) > 10 and ( re.search(r"[A-Za-z]", password) and re.search(r"\d", password) or re.search(r"[^A-Za-z0-9]", password) ):
        strength = "Strong"

    else:
        strength = "Medium"

    return jsonify({
        "password": password,
        "strength": strength
    })

# Run Flask App
if __name__ == '__main__':
    app.run(host='0.0.0.0',port = 5000, debug=True)
````

# **Problem Statement 9: Calculator API**

Imagine you are developing a basic calculator API that allows users to perform arithmetic operations directly from the browser.

## ✅ Your Goal:

* Create a Flask API in **GitHub Codespaces**.
* Create a home route `/` with instructions.
* Create an API route:

```text
/calculate/<num1>/<num2>
```

The API should return:

* Sum

* Difference

* Multiplication

* Division

* The API should work with any two numbers entered in the URL.

* If the second number is `0`, return:

```text
Cannot divide by zero
```

image.png

````python
#Problem Statement 9: Calculator API
from flask import Flask, jsonify

app = Flask(__name__)

# Home route
@app.route('/')
def home():
    return "Welcome to the Calculator API! Use /calculate/<num1>/<num2> to perform calculations."

# Calculator API
@app.route('/calculate/<float:num1>/<float:num2>', methods=['GET'])
def calculate(num1, num2):
    result = {
        "number_1": num1,
        "number_2": num2,
        "sum": num1 + num2,
        "difference": num1 - num2,
        "multiplication": num1 * num2
    }

    if num2 == 0:
        result["division"] = "Cannot divide by zero"
    else:
        result["division"] = num1 / num2

    return jsonify(result), 200

# Run the Flask app
if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000, debug=True)
````

````python
# Random Movie Suggestion API
from flask import Flask, jsonify
import random

app = Flask(__name__)

# List of movies
movies = [
    "Inception",
    "Interstellar",
    "Avengers: Endgame",
    "The Dark Knight",
    "3 Idiots",
    "Jawan",
    "Spider-Man: No Way Home",
    "Bahubali",
    "Titanic",
    "The Lion King"
]

# Home route
@app.route('/')
def home():
    return "Welcome to the Movie Suggestion API! Use /movie to get a random movie suggestion."

# Random movie API
@app.route('/movie', methods=['GET'])
def suggest_movie():
    random_movie = random.choice(movies)

    return jsonify({
        "movie_suggestion": random_movie
    })

# Run the Flask app
if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000, debug=True)
````
