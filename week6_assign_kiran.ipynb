1.Problem Statement:
Filter out names from a list that are longer than 5 characters.

Input: ['Alice', 'Bob', 'Catherine', 'David']

Expected Output: ['Catherine', 'David']

```python
#filster the names longer than 5 chars
names = ['Alice', 'Bob', 'Catherine', 'David']
filtered_names = list(filter(lambda x: len(x) >= 5, names))
print(filtered_names)
```

8.Problem Statement:
Design a Python program with an Inventory class to manage the stock of products. The class should allow updating the stock for products and displaying the current stock levels. Ensure the functionality is clear and concise.

Expected Output:

Stock updated for Apple: 50

Stock updated for Banana: 30

Current stock: {'Apple': 50, 'Banana': 30}

```python
#inventory class to manage the stock of products
class Inventory:
    def __init__(self):
        self.stock = {}
        #self.stock = {'Apple': 50, 'Banana': 30}
    def update_stock(self, product, quantity):
        if product in self.stock:
            self.stock[product] += quantity
        else:
            self.stock[product] = quantity
            #self.stock = {'Apple': 50, 'Banana': 30}
    def display_stock(self):
        return self.stock
        #return {'Apple': 50, 'Banana': 30}

#inventory object stock , update or display the stock
inventory_obj = Inventory()
inventory_obj.update_stock('Apple', 50)
inventory_obj.update_stock('Banana', 30)
print(f"Current Stock is: {inventory_obj.display_stock()}")
```

9.Problem Statement:
Create a Loan class that evaluates if a person qualifies for a loan based on their salary and the loan amount requested. The class should have methods to check if the loan can be approved or rejected based on simple criteria.

Expected Output:

Loan approved for $5000.

Loan rejected for $10000.

-- Attributes:

* salary: The person's salary.
* max_loan: Set to twice the salary (loan limit).

--Methods:

* check_loan: Checks if the requested loan amount is less than or equal to max_loan. If so, the loan is approved; otherwise, it is rejected.

-- Code Walkthrough:
* The Loan class is initialized with the person's salary.
* The max_loan is calculated as twice the salary.
* The check_loan method approves or rejects the loan based on the loan amount compared to max_loan.

Example:
* For a salary of $ 4000, the maximum loan is 8000.
* A 5000 loan is approved (within the limit).
* A 10000 loan is rejected (exceeds the limit).

```python
#loan class to evaluate if a person is eligible for a loan or not
class Loan:
    def __init__(self, salary):
        self.salary = salary
        self.max_loan = 2 * salary

        #check loan method
    def check_loan(self, loan_amount):
        if loan_amount <= self.max_loan:
            print(f"Loan approved for ${loan_amount}.")
        else:
            print(f"Loan rejected for ${loan_amount}.")

loan_obj = Loan(4000)
loan_obj.check_loan(5000)
loan_obj.check_loan(10000)
```

**4.Problem Statement:**
Write a lambda function to check if a given string is a palindrome. The lambda function should return True if the string is the same when read forwards and backwards, and False otherwise.

**Expected Output:**If the string is a palindrome, the output should be True, otherwise False.

**Example:**

Input: "racecar" Output: True
* "racecar" is a palindrome because if you read it from left to right or right to left, it’s the same.

Input: "hello" Output: False
* hello" → when read from left to right, it's "hello", but when read backward, it's "olleh". This is not a palindrome.

**Explanation:**A palindrome is a word or phrase that reads the same backward as forward, ignoring spaces, punctuation, and capitalization.


```python
#CHECKING IF THE GIVEN STRING IS PALINDROME OR NOT USING LAMBDA FUNCTION
is_palindrome = lambda s: True if s == s[::-1] else False
result = is_palindrome("racecar")
print(result)

is_palindrome = lambda s: True if s == s[::-1] else False
result = is_palindrome("HELLO")
print(result)
```
