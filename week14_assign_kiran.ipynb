14.Using a Foreign Key in SQL

Problem Statement:

A school wants to keep track of course enrollments. Each student can enroll in multiple courses, so we need a new Enrollments table where StudentID references the Students table as a Foreign Key .

Expected Output:

The Enrollments table is created with StudentID as a foreign key, linking it to the Students table.

```python
conn = sqlite3.connect('school.db')
cursor = conn.cursor()
cursor.execute("PRAGMA Foreign_key = ON")
conn.commit()
cursor.execute('''
CREATE TABLE IF NOT EXISTS Enrollments (
    EnrollmentID INTEGER PRIMARY KEY,
    StudentID INTEGER,
    CourseName TEXT,
    FOREIGN KEY (StudentID) REFERENCES Students(StudentID)
);
''')

conn.commit()
conn.close()
```

15.Truncating a Table (Deleting All Records Without Dropping the Table)

Problem Statement:

The company wants to remove all employee records from the Employees table while keeping the table structure intact.

Expected Output:

All records are deleted from the Employees table, but the table structure remains intact.

note: SQLite does not support TRUNCATE we perform Delete instead




```python
conn = sqlite3.connect('company.db')
cursor = conn.cursor()

cursor.execute('DELETE FROM Employees;')  # SQLite does not support TRUNCATE
conn.commit()

# Verify truncation
cursor.execute('SELECT * FROM Employees;')
remaining_records = cursor.fetchall()
for record in remaining_records:
    print(record)

conn.close()
```
