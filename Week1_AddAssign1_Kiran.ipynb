Problem Statement4: Develop a Python program that:

image.png

Problem Statement7: Write a Python program that asks the user for their name and age and convert the age into integer datatype, then calculates the year they will turn 100 years old and displays a message with that information.

ex: current_year + (100 - age)

Expected output:

image.png

```python
#enter your name
name=input("enter your name: ")

#enter your age
age=input("enter your age: ")

#convert age into integer datatype
age_int=int(age)

#enter current year
current_year=2026

#calculate the year they will turn 100
hundredth_year=current_year+(100-age_int)

#displaying the message
print(f"{name}, you will turn 100 years old in the year {hundredth_year}")
```
