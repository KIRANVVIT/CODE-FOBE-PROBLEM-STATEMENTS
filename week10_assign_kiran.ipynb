```python
#creating a dataset and converting it into dataframe
import pandas as pd
data = {
    'Month': ['January', 'January','February','February', 'March','March'],
    'Household_ID': ['H001', 'H002', 'H001', 'H003', 'H002', 'H004'],
    'Electricity_Bill': [120.50, 95.30, 110.40, 140.60, 100.00, 130.00],
    'Water_Bill': [35.20, 40.10, 34.80, 30.00, 45.50, 25.00],
    'Internet_Bill': [50.00, 55.00, 50.00, 60.00, 55.00, 65.00],
    'Total_Bill': [205.70, 190.40, 195.20, 230.60, 200.50, 220.00]
}
df = pd.DataFrame(data)
df
```

```python
#Calculate the total bills for each household (Household_ID) by summing up the Total_Bill across all months by using groupby.
#Group data by Household_ID and calculate the total bills
print(data_csv)
total_bills = data_csv.groupby('Household_ID')['Total_Bill'].sum()

#Display the total bills for each household
print(total_bills)
```
