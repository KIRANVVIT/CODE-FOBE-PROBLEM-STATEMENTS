1.Problem Statement:
* Two DataFrames contain employee details:
* df1: Employees from Department A with ['ID', 'Name', 'Department'].
* df2: Employees from Department B with the same columns.

Combine these two DataFrames vertically to create a single employee database.

image.png

```python
'''Two DataFrames contain employee details:
df1: Employees from Department A with ['ID', 'Name', 'Department'].
df2: Employees from Department B with the same columns.
Combine these two DataFrames vertically to create a single employee database.
'''
import pandas as pd
df1 = pd.DataFrame({
    'ID': [1, 2],
    'Name': ['Alice', 'Bob'],
    'Department': ['A', 'A']
    })
df2 = pd.DataFrame({
    'ID': [3, 4],
    'Name': ['Charlie', 'Dave'],
    'Department': ['B', 'B']
    })
employee_data = pd.concat([df1, df2], ignore_index=True)
print(employee_data)
```

2.Problem Statement:

You have two DataFrames:
* df1: Product names (['ProductID', 'Name']),
* df2: Categories for products (['ProductID', 'Category']).

Merge them to get product names with their categories.

image.png

```python
'''You have two DataFrames:
df1: Product names (['ProductID', 'Name']),
df2: Categories for products (['ProductID', 'Category']).
Merge them to get product names with their categories.
'''
df1 = pd.DataFrame({
    'ProductID': [1, 2, 3],
    'Name': ['Bread', 'Apple', 'Chocolate']
    })
df2 = pd.DataFrame({
    'ProductID': [1, 2, 3],
    'Category': ['Bakery', 'Fruits', 'Sweets']
    })

df = pd.merge(df1, df2, on='ProductID')
print(df)
```

4.Problem Statement:
* df1: Student names (['ID', 'Name']).
* df2: Test scores (['ID', 'Score']). Some students didn’t take the test.

Perform a left join to include all students and their scores.

image.png

```python
'''df1: Student names (['ID', 'Name']).
 df2: Test scores (['ID', 'Score']). Some students didn’t take the test.
Perform a left join to include all students and their scores.
'''
import pandas as pd
import numpy as np
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
    })
df2 = pd.DataFrame({
    'ID': [1, 2],
    'Score': [85, 90]
    })
df = pd.merge(df1, df2, on='ID', how='left')
print(df)

```
