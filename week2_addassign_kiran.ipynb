**5. Leap Year Checker**

Write a program that checks if a given year is a leap year. Use the input() function to take the year as input from the user.

A year is a leap year if it meets the following conditions:

* The year is evenly divisible by 4.
* If the year is also divisible by 100, then it must also be divisible by 400 to be a leap year.

Based on these rules:(use if and else )

* If the year is a leap year, the program should print: "<year> is a leap year".
* If the year is not a leap year, the program should print: "<year> is not a leap year".

**Example**

**Example Input:** year = 2020

**Expected Output:** 2020 is a leap year

**Explanation:**
2020 is divisible by 4, and since it is not a century year (not divisible by 100), it is a leap year

```python
# leap year checker of the year
year = int(input("Enter the year: "))

#checking leap year conditions
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year")
```

**7. Movie Ticket Pricing Based on Age and Showtime**

Write a program to calculate the ticket price for a movie in rupees based on the person’s age and the time of the show (morning or evening). Use the input() function to take both the age and showtime as inputs from the user.

The ticket pricing rules are as follows:

Ticket Prices by Age Group:

* Child (12 years old or younger): ₹100
* Adult (between 13 and 59 years old): ₹240
* Senior (60 years or older): ₹160

Discount Based on Showtime:

* Morning Show: If the showtime is "morning", apply a 20% discount to the ticket price.
* Evening Show: If the showtime is "evening", there is no discount applied.

**Example**

Example Input:

age = 30, showtime = 'morning'

**Expected Output:**

Ticket Price: ₹192.0

**Explanation:**
In this example, the person is 30 years old, which falls into the adult category, so the base ticket price is ₹240. Since the showtime is in the morning, a 20% discount is applied:

Discounted Price = ₹240 - (20% of ₹240) = ₹192

If the showtime was "evening" for the same age, the ticket price would remain ₹240 as no discount applies.

```python
from IPython.core.interactiveshell import dis
#using input() to get age and showtime(morning/evening)
age = int(input("enter your age: "))
show_time = str(input("enter the showtime(morning/evening): "))

#princing based on age
if age <= 12:
  price = 100
elif 13 <= age <= 59:
  price = 240
else:
  price = 160

#discount based on show time
if show_time == "morning":
  disc = 0.2
else:
  disc = 0

#final price
final_price = price - (price*disc)

print(f"Ticket Price: Rs.{final_price}")
```

**Input Format:**

1. Total bill amount (float)

2. Service charge percentage (float)

3. Number of friends (integer)

**Output Format:**

Print “Each pays: <amount>” (rounded to 2 decimals)

**Example Input:**

**11. Gym Membership Discount Calculator**

A gym offers membership discounts based on age and joining time. Younger or senior members get special offers, and morning joiners get an extra discount.

Write a program to calculate the final membership fee after applying discounts.

**Rules:**

1. Base fee = ₹1500

2. If age < 21 or age ≥ 60 → 15% off

3. If joining in the morning → extra 5% off on the discounted price

**Input Format:**

1. Age (int)

2. Time of joining (string: morning/evening)

**Output Format:**

Final fee rounded to 2 decimals.

**Example Input:**
1. Enter age: 19
2. Enter joining time (morning/evening): morning

**Expected Output:**

Final fee: 1211.25

```python
#gym membership discount calculator
#enter int age and joining time string morning or evening
base_fee=1500
age=int(input("Enter age: "))
joining_time=str(input("Enter joining tiem(morning/evening): "))

#applying gym rules
if age<21 or age>=60:
  discount=15
  base_fee=base_fee-(base_fee*(discount/100))

if joining_time=="morning":
  discount=5
  base_fee=base_fee-(base_fee*(discount/100))

#printing final fee
print(f"Final fee: {base_fee:.2f}")
```
