**Problem Statement2:**Create a Python program that takes three numbers as input from the user, stores them in three separate variables and converts the input values to floating-point numbers using float() to handle both integers and decimals , and checks which number is the largest among the three. Use comparison operators (>=) along with the and operator to ensure the correct logic for identifying the largest number. If two or more numbers are equal and also the largest, the program should still print that number. Display the result using a formatted string.

Example Input and Output:

Input:

Enter the first number: 10

Enter the second number: 20

Enter the third number: 15

Output:The largest number is: 20.0

```python
#finding the largest of 3 numbers
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
num3 = float(input("Enter the third number: "))
if num1 >= num2 and num1 >= num3:
  print(f"The largest number is: {num1}")
elif num2>=num1 and num2>=num3:
  print(f"The largest number is: {num2}")
else:
  print(f"The largest number is: {num3}")
```

**Problem Statement3:**Create a Python program that takes the total bill amount as input from the user. If the total bill is greater than ₹1000, the program should calculate a 10% discount by multiplying the total bill by 0.1 (10%) and subtracting the discount from the total bill. If the bill is ₹1000 or less, no discount is applied. The program should print the final amount using a formatted string.

Example Input and Output:

Input:

Enter the total bill amount: ₹1200

Output:

Discount applied. Final amount: ₹1080.0

```python
#billing the amount and applying discount
bill = float(input("Enter the total bill amount: ₹"))
if bill > 1000:
  discount = bill * 0.1
  final_amount = bill-discount
else:
  final_amount = bill
print(f"Discount applied. Final amount: ₹{final_amount}")
```

**Problem Statement5:**Create a program that takes three numbers as input from the user and converts the input values to floating-point numbers using float() to handle both integers and decimals, stores them in three separate variables, calculates their average by adding all the three numbers and then divide by 3, and prints the result using a formatted string.

Example Input:

Enter the first number: 10.0

Enter the second number: 20.0

Enter the third number: 30.0

**Example Output:**The average of 10.0, 20.0, and 30.0 is: 20.0

```python
#finding average of 3 numbers
value1=float(input("Enter the first number: "))
value2=float(input("Enter the second number: "))
value3=float(input("Enter the third number: "))
average=(value1+value2+value3)/3
print(f"The average of {value1}, {value2} and {value3} is: {average}")
```
