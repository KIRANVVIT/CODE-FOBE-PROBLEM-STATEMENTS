**Problem1:** Write a program to simulate the roll of a six-sided die until the outcome is a six using random.randint. Each time the die is rolled, print the result. Continue rolling and printing the outcome until a six appears. Use a while loop to perform repeated rolls and stop once the result is six.

**Expected Output:**

Rolled: 3

Rolled: 5

Rolled: 6

(Note: Since the rolls are random, the output values will vary each time the program is run.)

```python
# rolling the die until we get 6
import random

while True:
    num = random.randint(1, 6)
    if num == 6:
        break
    print("Rolled:", num)
print("Rolled: 6")
```

**Problem5:** Write a program that prompts the user to enter a sentence using the input() function. Then, count and print the number of vowels (both uppercase and lowercase) present in the sentence. Vowels are defined as the letters 'a', 'e', 'i', 'o', and 'u', and the program should recognize both uppercase and lowercase versions of these vowels. Use a for loop to iterate through each character in the sentence, and an if statement to check if the character is a vowel.

**Expected Output:**

Enter a sentence: Hello World

Vowel count: 3

(Note: The vowel count may vary depending on the sentence entered by the user.)

```python
#counting the no.of vowels in the given sentence
sentence = input("Enter a sentence: ")
vowel_count = 0
for char in sentence:
    if char.lower() in "aeiou":
        vowel_count += 1
print("Vowel count:", vowel_count)
```

**Problem6:** Write a program that prompts the user to enter 5 names, one at a time, and store each name in a list. Start by creating an empty list to hold the names. Use the input() function to ask for each name and append it to the list. After all names are entered, print the list of names in reverse order. To reverse the list, use list slicing ([::-1]). This will display the names in the opposite order from how they were entered.

**Expected Output:**

Enter name 1: Alice

Enter name 2: Bob

Enter name 3: Charlie

Enter name 4: David

Enter name 5: Eve

Names in reverse: ['Eve', 'David', 'Charlie', 'Bob', 'Alice']

```python
# list operations
names_list = []
for i in range(1,6):
    name = input(f"Enter name {i}: ")
    names_list.append(name)
#printing the names list in reverse order using slicing [:-1]
print("Names in reverse:", names_list[::-1])
```

**Problem7:** Write a program that asks the user to enter 5 temperatures in Celsius, one at a time. Store each Celsius temperature in a list. After collecting all 5 temperatures, convert each temperature to Fahrenheit using the formula F = C × 9/5 + 32 where C is the temperature in Celsius. Print both the original Celsius temperatures and the converted Fahrenheit temperatures as two separate lists.

* Use the input() function to collect each temperature from the user.
* Use a for loop to repeat the input collection 5 times.
* After collecting all inputs, create a new list where each temperature in Celsius is converted to Fahrenheit.

Formula for Conversion: F = C × 9/5 + 32

Expected Output:

Enter temperature 1 in Celsius: 25

Enter temperature 2 in Celsius: 30

Enter temperature 3 in Celsius: 20

Enter temperature 4 in Celsius: 15

Enter temperature 5 in Celsius: 10

Celsius: [25.0, 30.0, 20.0, 15.0, 10.0]

Fahrenheit: [77.0, 86.0, 68.0, 59.0, 50.0]

(Note: The specific output values may vary depending on the user inputs.)

```python
# temperature conversion from celcius to  fahrenheit
celcius = []
fahrenheit = []
for i in range(1,6):
    temp = float(input(f"Enter temperature {i} in Celsius: "))
    celcius.append(temp)
print ("\nCelsius:", celcius)
for i in celcius:
    fahrenheit.append(i*9/5+32)
print ("\nFahrenheit:", fahrenheit)
```
