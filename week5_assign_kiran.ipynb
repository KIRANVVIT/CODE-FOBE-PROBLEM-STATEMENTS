Problem 8 : Delete a Line by Index from a File
Problem Statement: Delete a specific line (e.g., the third line) from the file user_activity.txt. After deletion, update the file to reflect the changes using with statement.

Expected Output:

Updated content of the user_activity.txt file:

User logged in at 09:00 AM

User viewed product details at 09:15 AM

User logged out at 10:00 AM

User added item to wishlist at 10:30 AM

```python
#deleting the 3rd line in the file and update the file using with stmt
with open("user_activity.txt", "r") as file:
    lines = file.readlines()
    del lines[2]

with open("user_activity.txt", "w") as file:
    file.writelines(lines)

print("Updated content of the user_activity.txt file:\nafter deleting 3rd line\n")
for line in lines:
    print(line)
```

Problem 9 : Create and Write to a CSV File with User Data
Problem Statement: Create a CSV file called users.csv that contains a list of users. Each user has a name, age, and email. Write this information to the CSV file using with statement.

Expected Output:

CSV file content:

John Doe,28,john.doe@example.com

Jane Smith,34,jane.smith@example.com

Bob Johnson,45,bob.johnson@example.com

```python
#creating csv file of users with their name,age and email using with stmt
import csv

with open("users.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Sr.No", "Name", "Age", "Email"])
    writer.writerow(["1","John Doe", "28", "john.doe@example.com"])
    writer.writerow(["2","Jane Smith", "34", "jane.smith@example.com"])
    writer.writerow(["3","Bob Johnson", "45", "bob.johnson@example.com"])

#reading the content of the file
with open("users.csv", "r") as file:
    content = file.read()

print("Content of the CSV file is:\n")
print(content)
```

1.Problem:Create a function named convert_to_fahrenheit. This function should take one argument, a temperature in Celsius, and then convert that temperature to Fahrenheit using a formula. Finally, it should return the Fahrenheit temperature.

Background: The formula to convert Celsius to Fahrenheit is:
Fahrenheit = Celsius x 9/5 + 32

Steps
* Define a function called convert_to_fahrenheit.
* The function should accept one input parameter, celsius, representing a temperature in Celsius.
* Inside the function, use the conversion formula to calculate the Fahrenheit equivalent of celsius.
* Return the resulting Fahrenheit temperature.

Expected Output:

convert_to_fahrenheit(100)

Output: 212.0

```python
#1.Convert to fahrenheit function
#defining the function
def convert_to_fahrenheit(celsius):
    fahrenheit = celsius * 9/5 + 32
    return fahrenheit

#calling the function
convert_to_fahrenheit(100)
```

3.Problem: Write a function named is_even that checks if a given number is even.

Steps
* Define the Function: Create a function called is_even.
* Parameter: The function will take one parameter:
* number: an integer that you want to check.
* Determine Evenness: Use the modulo operator % to check if number is divisible by 2. If number % 2 equals 0, the number is even; otherwise, it is odd.
* Return Result: If the number is even, the function should return True; if it’s odd, return False.

Expected Output:

Test Case 1

print(is_even(4))

* Expected Output: True
* Explanation: 4 is even, so the function returns True.

```python
#3. function for checking if the given num is even or not
#defining a function
def is_even(number):
    if number % 2 == 0:
        return True
    else:
        return False

#calling function
print(is_even(4))
```

9.Problem Statement: Create a function reverse_string that accepts a string and returns its reversed version using [::-1].

Expected Output:

For the input "hello":

olleh

```python
#9. reverse a string function
def reverse_string(string):
    reversed_string = string[::-1]
    return reversed_string

#calling function
reverse_string("hello")
```
