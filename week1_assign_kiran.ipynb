**Problem Statement 5**:Write a program that defines a boolean variable passed_exam to indicate whether you passed an exam. Convert this boolean value to a string and print it in a sentence. Also, display the data types of the variable both before and after conversion using the type() function..

Example Output:

Did I pass the exam? True

Data type of passed_exam (before conversion): <class 'bool'>

Data type of passed_exam_str (after conversion): <class 'str'>

```python
#defining a boolean variable of exam status
passed_exam=True

#conversion of boolean variable into string
passed_exam_str=str(passed_exam)

#printing the status of exam and the datatypes of boolean and string variables
print(f"Did I pass the exam? {passed_exam}")
print("Data type of passed_exam (before conversion): ",type(passed_exam))
print(f"Data type of passed_exam_str (after conversion): {type(passed_exam_str)}")
```
