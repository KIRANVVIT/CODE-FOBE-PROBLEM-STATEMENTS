13.Find Students Above Average Marks

Problem: Marks of students are [45, 78, 56, 89, 74, 92]. Find all students who scored above the average.



Expected Output:


Average Marks: 72.33333333333333

Above Average Marks: [78 89 74 92]

Explanation: Calculate the average using mean(). Filter values greater than the average.



```python
#students above average marks
marks = np.array([45, 78, 56, 89, 74, 92])
average = np.mean(marks)
print("Average Marks:", average)
above_average = [int(x) for x in list(filter(lambda x : x > average, marks))]
print("Above Average Marks:", above_average)
```
