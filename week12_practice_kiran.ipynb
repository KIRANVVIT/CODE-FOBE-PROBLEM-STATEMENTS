```python
'''Define the data:
Days: [1, 2, 3, 4, 5, 6, 7] (Monday to Sunday).
Sales: [50, 60, 70, 65, 80, 90, 100] (cups of coffee sold).
Create the line plot:
Connect the points to show the sales trend over the week.
Interpretation: Identify peak sales (Saturday and Sunday) and slow days (Monday).
                                      line plot
'''
import matplotlib.pyplot as plt
#Data
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
coffee_sales = [40, 50, 60, 70, 80, 90, 100]  # default solid line, color red , markers circle o
sales1 = [45, 55, 65, 75, 85, 95, 105]     # dashed line  , blue  , square s
sales2 = [50, 60, 70, 80, 90, 100, 110]     # dotted line  , green  , triangle ^
sales3 = [60, 70, 80, 90, 100, 110, 120]     # dash dot line , black , star *

# Create the line plot
plt.plot(days, coffee_sales, marker='o', color='red', label = 'Coffee Sales')# Label added
plt.plot(days, sales1, marker='s', color='blue', linestyle='--' , label = 'Sales1')# Label added
plt.plot(days, sales2, marker='^', color='green', linestyle=':' , label = 'Sales2')# Label added
plt.plot(days, sales3, marker='*', color='black', linestyle='-.', label = 'Sales3')# Label added

# Customize the plot
plt.title('Coffee Sales Over the Week')
plt.xlabel('Days')
plt.ylabel('Cups Sold')
plt.legend()  # Add legend to differentiate lines
plt.legend(loc="upper right", bbox_to_anchor=(1.3, 1))  # Add legend to differentiate lines
# Show the plot
plt.grid(True)
plt.show()
```

```python
''' Comparing values across categories (e.g., sales, profits, scores).
compare the monthly sales revenue of your top 4 products: Coffee, Tea, Juice, and Water.
                  Matplotlib Bar Plot
'''
import matplotlib.pyplot as plt
# Data
categories = ["Coffee", "Tea", "Juice", "Water"]
sales = [1200, 850, 400, 500]
# Plotting
plt.bar(categories, sales,color= "yellow", width=0.5, edgecolor='red')
# Adding labels and title
plt.title('Monthly Sales Revenue',fontsize=16, fontweight="bold", color='darkblue')
plt.xlabel('Products',fontsize=12)
plt.ylabel('Sales Revenue',fontsize=12)
# Adding grid lines
plt.grid(axis="y", linestyle="--", color='black', alpha=0.7)# alpha is visibility of grid line, range 0-1

plt.show()
```

```python
#                                    horizontal bar
import matplotlib.pyplot as plt
# Data
categories = ["Coffee", "Tea", "Juice", "Water"]
sales = [1200, 850, 400, 500]
# Plotting
plt.barh(categories, sales,color= ["fuchsia","green","blue","red"], height=0.5, edgecolor='yellow')
# Adding labels and title
plt.title('Monthly Sales Revenue',fontsize=16, fontweight="bold", color='darkblue')
plt.xlabel('Products',fontsize=12)
plt.ylabel('Sales Revenue',fontsize=12)
# Adding grid lines
plt.grid(axis="x", linestyle="--", color='black', alpha=0.7)# alpha is visibility of grid line, range 0-1

plt.show()
```

```python
#                                HISTOGRAM
import matplotlib.pyplot as plt
#Data
weights = [50, 55, 60, 62, 63, 65, 68, 70, 72, 72, 75, 78, 80, 82, 85, 90, 92, 95]
# Plotting
plt.hist(weights, bins=5, color="fuchsia", edgecolor="red")
# Adding labels and title
plt.title("Distribution of Client Weights", fontsize=16, color="darkblue")
plt.xlabel("Weight Range (kg)", fontsize=12)
plt.ylabel("Number of Clients", fontsize=12)
# Show the plot
plt.show()
```

```python
#                              PIE CHART, Customized Pie Chart
import matplotlib.pyplot as plt
# Data
menu_items = ["Pizza", "Pasta", "Burgers", "Drinks"]
revenue = [40, 25, 20, 15]
# Plotting
plt.pie(revenue, labels=menu_items, autopct="%1.2f%%")
# Adding title
plt.title("Revenue Breakdown by Menu Items")
# Show the plot
plt.show()
```

```python
#              CUSTOMISED PIE CHART
import matplotlib.pyplot as plt

# Data
categories = ["Rent", "Groceries", "Utilities", "Entertainment"]
expenses = [50, 25, 15, 10]
colors = ["gold", "lightgreen", "lightblue", "coral"]
explode = [0, 0, 0, 0.1]  # Highlight Rent slice
# Figure size
plt.figure(figsize=(20, 5))
plt.pie(expenses, labels=categories, autopct='%1.1f%%', startangle=140, colors=colors, explode=explode, shadow=True)
# Adding legend
plt.legend(loc="upper right", bbox_to_anchor=(1.6, 1))  # Move legend outside the chart
# Adding title
plt.title("Household Monthly Expenses Breakdown", fontsize=16, fontweight="bold")

# Show the plot
plt.show()
```
